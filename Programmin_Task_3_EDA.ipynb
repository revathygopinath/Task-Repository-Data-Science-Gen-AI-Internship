#1. Maximum products of two elements in an array
Given the array of integers nums, you will choose two different indices i and j of that array. Return the maximum value of (nums[i]-1)*(nums[j]-1).



```python
class Solution(object):
  def MaxProduct(self,nums):
    nums.sort()
    largest_num = nums[-1]
    sec_largest = nums[-2]
    return (largest_num -1) *(sec_largest-1)
```

```python
obj = Solution()
result = obj.MaxProduct([3,4,5,2])
print(result)
```

#2. Count number of teams

```python
class Solution(object):
    def numTeams(self, rating):
        n = len(rating)
        teams = 0

        for j in range(n):
            left_small = 0
            left_large = 0
            right_small = 0
            right_large = 0

            # check left side of j
            for i in range(j):
                if rating[i] < rating[j]:
                    left_small += 1
                else:
                    left_large += 1

            # check right side of j
            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_large += 1
                else:
                    right_small += 1

            teams += left_small * right_large
            teams += left_large * right_small

        return teams
```

```python
obj = Solution()
result =obj.numTeams([2,1,3])
print(result)
```

#3. Number of students doing home work at a given time
Input: startTime = [1,2,3], endTime = [3,2,7], queryTime = 4
Output: 1

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
      result =0
      for x,y in zip(startTime,endTime):
        if min(x,y) <=queryTime <=max(x,y):
          result = result+1
      return result
```

```python
obj = Solution()
result = obj.busyStudent([1,2,3],[3,2,7],4)
print(result)
```

#4. Number of steps to reduce a number to Zero.
Input: num = 14
Output: 6

```python
class Solution(object):
    def numberOfSteps(self, num):
        result =0
        while num !=0:
          if num % 2==0:
            num = num//2
          else:
            num = num-1
          result =result+1
        return result
```

```python
obj = Solution()
result =obj.numberOfSteps(14)
print(result)
```

#5. Counting Bits
Example 1:

Input: n = 2
Output: [0,1,1]
Explanation:
0 --> 0
1 --> 1
2 --> 10

```python
class Solution(object):
    def countBits(self, n):
      result =[]
      for i in range(n+1):
        binary = bin(i)
        count = binary.count('1')
        result.append(count)
      return result
```

```python
obj = Solution()
result = obj.countBits(5)
print(result)
```
