#1. Defanging an IP Address

```python

class Solution(object):
    def defangIPaddr(self, address):
        return address.replace(".", "[.]")
```

```python
sol =Solution()
result = sol.defangIPaddr("1.1.1.1")
print(result)
```

```python
# Using Loop
class solution(object):
  def defangIPaddr(self,address):
    result =""
    for i in address:
      if i ==".":
        result = result +"[.]"
      else:
        result =i
    return result
```

#2. Find numbers with an even number of digits

```python

class Solution1(object):
    def findNumbers(self, nums):
        result = 0
        for i in nums:
            if len(str(i)) % 2 == 0:
                result = result + 1
        return result
```

```python
sol1 = Solution1()
result = sol1.findNumbers([12,345,2,6,7896])
print(result)
```

#3. Number of Good Pairs

```python

class Solution2(object):
    def numIdenticalPairs(self, nums):
        result =0
        n =len(nums)
        for i in range(n):
          for j in range(1+i,n):
            if nums[i] ==nums[j]:
              result = result+1
        return result
```

```python
sol = Solution2()
obj = sol.numIdenticalPairs([1,2,3,1,1,3])
print(obj)
```

#4. How many numbers are smaller than the current number

```python

class Solution3(object):
  def smallerNumbersThanCurrent(self,nums):
    result =[]
    for i in nums:
      count =0
      for j in nums:
        if i>j:
          count=count+1
      result.append(count)
    return result


```

```python
sol =Solution3()
result = sol.smallerNumbersThanCurrent([8,1,2,2,3])
print(result)
```

#5. Substract the product and sum of digits of an integer

```python


class Solution4(object):
    def subtractProductAndSum(self, n):
      product =1
      sum =0
      for i in str(n):
        sum = sum + int(i)
        product = product*int(i)
      result = product -sum
      return result
```

```python
sol = Solution4()
obj = sol.subtractProductAndSum(569)
print(obj)
```

#6. XOR Operation in an array

```python
class Solution6(object):
    def xorOperation(self, n, start):
        result = 0

        for i in range(n):
            num = start + 2 * i
            result = result ^ num
        return result
```

```python
obj = Solution6()
result = obj.xorOperation(4,3)
print(result)
```
